```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension()
```

```python
df = pd.read_csv('Shobot_data.csv',index_col=0)
```

```python
df
```

```python
bcno = pn.widgets.TextInput(name='BC number',placeholder='Write the barcode number')

display(pn.Row(bcno, height=100))
```

```python
st = pn.widgets.StaticText(name = 'Name',value = 'Default')
```

```python
st1 = pn.widgets.StaticText(name='Weight',value='Default')
```

```python
st2 = pn.widgets.StaticText(name='Price',value='Default')
```

```python
st3 = pn.widgets.StaticText(name='Expiry Date',value='Default')
```

```python
def output(bcno):
    temp =df.loc[df['Barcode number']==bcno]
    if temp.empty:
        st.value ="No such item"
        st1.value ="No such item"
        st2.value ="No such item"
        st3.value ="No such item"
    else:
        st.value = temp.iloc[0,1]
        st1.value = str(temp.iloc[0,2]) +"g"
        st2.value = "Rs."+str(temp.iloc[0,3])
        st3.value = temp.iloc[0,4]
        
```

```python
output('8 901499 008176')
```

```python
bond =pn.bind(output,bcno)
```

```python
layout = pn.WidgetBox(bcno,bond,st,st1,st2,st3,align="center")
```

```python
layout
```

```python
layout.servable()
```

```python
template = pn.template.BootstrapTemplate(
    title='Shobot database',
    sidebar=[bcno],
)
```

```python
template.main.append(
    pn.Column(
        bond,st,st1,st2,st3,height = 500
    )
)
```

```python
template.servable()
```

```python
pn.serve(template)
```
